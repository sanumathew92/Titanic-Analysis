```python
# Titanic Dataset Analysis

# Step 1: Import Libraries
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt


# Step 2: Load Dataset
df = pd.read_csv("titanic-1.csv")
```

```python
# Step 2: Data Exploration and Cleaning
print("First 5 Rows:")
print(df.head())
```

```python
print("\nDataset Info:")
print(df.info())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Handle Missing Values
# Fill 'age' with median, 'embarked' with mode, drop 'deck' (too many missing) 
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop(columns=['Cabin'])
```

```python
# Print("\nMissing Values after Cleaning:")
print(df.isnull().sum())
```

```python
# Step 3: Data Analysis

# Survival Rate
survival_rate = df['Survived'].mean() * 100
print(f"\nOverall Survival Rate: {survival_rate:.2f}%") 


# Passenger Class Analysis
pclass_survival = df.groupby('Pclass')['Survived'].mean() * 100
print(pclass_survival) 

# Gendewr Analysis
gender_survival = df.groupby('Sex')['Survived'].mean() * 100
print("\nSurvival Rate by Gender:")
print(gender_survival)

# Age Analysis: Create age groups
df['age_group'] = pd.cut(df['Age'], bins=[0,12,18,30,50,80],
                         labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])
age_survival = df.groupby('age_group')['Survived'].mean() * 100
print("\nSurvival Rate by Age Group:")
print(age_survival)
```

Analysis:

1) The Overall Survival rate by 38%
2) This means most passengers did not survive the tragedy
3) 1st Class passengers had the highest survival rate (over 60%)
4) 3rd Class passengers had the lowest survival rate (below 25%)
5) This means social and economic inequality affected survival chances.
6) Women had a much higher survival rate(74%) compared to men (19%)
7) This reflects the "Women and Children First" policy followed towards during evacuation.
8) Children (0-12 years) had better survival chances than adults.
9) Young Adults(18-30 years) had the lowest survival rate
10) Older Adults (50+) survival rate dropped again, but not as low as young adult males.
11) Passengers who paid higher fares had higher survival rates
12) This means passengers who paid higher fares(wealthy travellers) had better access to lifeboats
13) 

```python
# Step 4: Data Visualization

plt.figure(figsize=(12,8))

# Age Distribution
plt.subplot(2,2,1)
sns.histplot(df['Age'], kde=True, bins=30, color='skyblue')
plt.title("Age Distribution") 

plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12,8))

# Fare Distribution 

plt.subplot(2,2,2)
sns.histplot(df['Fare'], kde=True,bins=30, color='orange')
plt.title("Fare Distribution") 

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

# Survival by Gender 

plt.subplot(2,2,3)
sns.barplot(x='Sex', y='Survived', data=df, palette="Set1")
plt.title("Survival Rate by Passenger Class")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

# Survival by Class

plt.subplot(2,2,4)
sns.barplot(x='Pclass', y='Survived', data=df, palette='Set1')
plt.title("Survival Rate by Passenger Class") 

plt.tight_layout()
plt.show()
```

```python
# Correlation HeatMap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm',fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
Correlation:

1) Survival is positively correlated with Fare and 1st Class
2) Male Gender and 3rd Class were strongly correlated with non-survival

Final Conclusion:
Survival in Titanic was not random
It was influenced by socioeconomic status, gender and age
First-class women and children had the best chances of survival
Third-class men had the worst chances of survival.

This shows how social hierarchy and evacuation policies shaped survival outcomes during the disaster.
```
