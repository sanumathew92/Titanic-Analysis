## In this exploratory data analysis project, we delve into the usage patterns of mobile devices across different demographics. With a dataset encompassing various metrics such as screen time, data usage, and e-commerce spending, we aim to uncover intriguing trends and insights. By leveraging Python libraries like Pandas, Matplotlib, and Seaborn, we seek to understand how different age groups interact with their devices and which mobile brands are favored across these groups. 



```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset

data = pd.read_csv('phone_usage_india.csv') ## Load dataset
data
```

```python
# Display Summary Statistics
data.shape #17686 rows × 16 columns
```

```python
data.describe()
```

```python
# Check for missing values

data.info()
```

```python
# Handle Missing, Incorrect, and Invalid Data

data.isnull().sum() #so there is no null/ missing values
```

```python
# creating a new column

data['Total screen & stream Time'] = data['Screen Time (hrs/day)'] + data['Streaming Time (hrs/day)']
data.head()
```

```python
# Explore Distributions of Numeric Columns Using Histograms

numeric_cols = ['Age', 'Screen Time (hrs/day)', 'Data Usage (GB/month)', 'Calls Duration (mins/day)', 'Number of Apps Installed', 'Social Media Time (hrs/day)', 'E-commerce Spend (INR/month)', 'Streaming Time (hrs/day)', 'Gaming Time (hrs/day)', 'Monthly Recharge Cost (INR)'] 

data[numeric_cols].hist(bins=15, figsize=(15, 10), color='skyblue') 
plt.suptitle('Distribution of Numeric Columns') 
plt.show()
```

```python
# Most Installed Apps

avg_apps_installed = data.groupby('Location')['Number of Apps Installed'].mean()
avg_apps_installed
```

```python
# Which mobile brand is most popular within different age categories

def categorize_age(age):     #Define age categories
    if age < 20:
        return "less than 20"
    elif 20 <= age < 30: 
        return '20-29' 
    elif 30 <= age < 40: 
        return '30-39' 
    elif 40 <= age < 50: 
        return '40-49' 
    elif 50 <= age < 60: 
        return '50-59' 
    else: 
        return '60 and above'

# Create new age category column
data['age category'] = data['Age'].apply(categorize_age)


# Group data by age category and mobile brand
age_barand_counts = data.groupby(['age category','Phone Brand']).size().unstack().fillna(0)
age_barand_counts
```

```python
ax = age_barand_counts.plot(kind='bar', stacked=True, figsize=(10, 6), cmap='tab20', title='Mobile Brand Distribution by Age Category') 
ax.set_xlabel('age category') 
ax.set_ylabel('Number of Users') 
plt.legend(title='Phone Brand', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Which location has the highest average screen time per day?

# Group by location and calculate the average screen time 
avg_screen_time = data.groupby('Location')['Screen Time (hrs/day)'].mean().sort_values(ascending=False) 
print(avg_screen_time) 

# Bar chart for average screen time by location 
avg_screen_time.plot(kind='bar', color='teal', title='Average Screen Time by Location')
plt.xlabel('Location') 
plt.ylabel('Average Screen Time (hrs/day)')
plt.show()
```

```python
# How does e-commerce spending vary by different phone brands?

# Group by phone brand and calculate the average e-commerce spend
avg_ecommerce_spend = data.groupby('Phone Brand')['E-commerce Spend (INR/month)'].mean().sort_values(ascending=False)
print(avg_ecommerce_spend)

# Bar chart for average e-commerce spend by phone brand
avg_ecommerce_spend.plot(kind='bar', color='coral', title='Average E-commerce Spend by Phone Brand')
plt.xlabel('Phone Brand')
plt.ylabel('Average E-commerce Spend (INR/month)')
plt.show() 
```
